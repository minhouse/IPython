# 最適化問題とは

関数 $f(\mathbf{x})=f(x_1,x_2,\ldots,x_n)$ が最小となる $\mathbf{x}=(x_1,x_2,\ldots,x_n)$を求める問題を**最適化問題(optimization problem)**と呼びます。例えば

$$ f(x,y) = x^2+2xy+3y^2\text{が最小となる $x,y$ を求めよ}$$

などが最適化問題の例です。関数の最大値を考えたい事もあると思いますが、数学的には関数 $f(\mathbf{x})$ を最大にする事と、 $-f(\mathbf{x})$ を最小にする事が同値なので、全て最小化問題として考える事が出来ます。

変数 $\mathbf{x}$ の取りうる値の範囲に制限が付いた物を**制約付き最適化問題(constrained optimization problem)**と呼びます。例えば

$$ \text{$x^2+y^2\leq 0$の時に $f(x,y) = x+2y$ が最小となる $x,y$ を求めよ} $$

という問題です。

## 何故最適化問題が重要か？
現実に現れる問題の多くは

> 最も良い～を求める

という形になります。例えば「最も利益の出る戦略を求めたい」「最も精度の良いパターン認識をしたい」「最も良く当たる未来予測をしたい」などです。
これを数学的に表現する為には、

1. 取りうる選択肢(パラメータ)に対応する変数 $\mathbf{x}=(x_1,x_2,\ldots,x_n)$ を決める
2. 「選択の良さ」を表現する関数 $f(\mathbf{x})$ を決める
3. $f(\mathbf{x})$ を最小化する $\mathbf{x}$ を求める

という形の最適化問題になります。このように非常に多くの問題を最適化問題として取り扱う事ができる為重要なのです。

## 最適化問題の表記
一々

> ～の時に～が最小となる～を求めよ

という形で日本語で書くと長くなりますし、自然言語は曖昧性があります。そこで、

$$ \text{条件 $C$ の下で関数 $f(\mathbf{x})$ が最小となる $\mathbf{x}$ を求めよ} $$

という問題を

$$ \underset{\mathbf{x}}{\text{minimize}}\ f(\mathbf{x})\ \text{subject to}\ C $$

とか

$$ \begin{array}{ccc}
\underset{\mathbf{x}}{\text{min}} & : & f(\mathbf{x}) \\
\text{sub.to} & : & C\\
\end{array}$$

などと書きます。変数がどれか明らかな時は，$\underset{\mathbf{x}}{\text{min}}$ の $\mathbf{x}$ を省略します。最大化問題を書くときも同様です。

## 最適化問題の種類

最適化問題は
1. 変数 $\mathbf{x}$ がどのようなものか
2. 変数 $\mathbf{x}$ の制約がどのようなものか
3. 関数 $f(\mathbf{x})$ がどのようなものか
によって種類が分かれます。1,2,3の組み合わせで最適化問題が決まるので、その種類は非常に多いです。それぞれについて適切な解法も異なってきます。

### 変数の種類による分類
まず、変数 $\mathbf{x}$ が実数や複素数などの連続変数である問題を**連続最適化問題 (continuous optimization problem)**と呼びます。上に挙げた2例は連続最適化問題です。一方， $\mathbf{x}$ が整数などの離散変数である問題を**離散最適化問題 (discrete optimization problem)**と呼びます。例えば

$$ \underset{n,m\in\mathbb{N}}{\text{minimize}}\quad f(n,m)=n^2+3m^2 $$

という問題が離散最適化問題の例です。$\mathbf{x}$ に連続変数と離散変数が混ざっている問題を**混合最適化問題 (mixed optimization problem)**と呼びます。

### 制約の種類による分類
等式の形の制約を**等式制約(equality constraint)**と呼びます。例えば

$$ \underset{x,y}{\text{minimize}}\quad x+2y\quad\text{subject to}\quad x^2+y^2 = 1 $$

が等式制約の例です。また、不等式の形の制約を**不等式制約(inequality constraint)**と呼びます。例えば

$$ \underset{x,y}{\text{minimize}}\quad x^2+2xy+3y^2\quad\text{subject to}\quad x\geq 0,y\geq 0$$

が不等式制約の例です。

また、制約を表現する式の形も重要です。
$$ a_0 + a_1x_1+a_2x_2 + \cdots + a_nx_n = 0 $$
や
$$ a_0 + a_1x_1+a_2x_2 + \cdots + a_nx_n \geq 0$$
のように一次式で書くことが出来る制約を**線形制約(linear constraint)**, そうでない制約を**非線形制約(nonlinear constraint)**と呼びます。
例えば、先ほどの $x^2+y^2=1$ は非線形等式制約、$x\geq 0,y\geq 0$ は線形不等式制約という事になります。

線形であれば、行列・ベクトルによって表現し線形代数の知識を使って効率の良い解法を作る事が可能なのでこの区別は非常に重要です。

### 関数の種類による分類
関数 $f(\mathbf{x})$ が $\mathbf{x}$ の一次式で書ける問題を**一次最適化問題(linear optimization problem)**と呼びます。
つまり，適当なベクトル $\mathbf{a}$ と定数 $c$ によって

$$ f(\mathbf{x}) = \mathbf{a}^T\mathbf{x}+c = \sum_i a_ix_i + c $$

と書けるのが一次最適化問題です。一次最適化問題以外は**非線形最適化問題(nonlinear optimization problem)**と呼びます。

$\mathbf{x}$ の二次式で書ける問題を**二次最適化問題(quadratic optimization problem)**と呼びます。
つまり，適当な正方行列 $\mathbf{A}$ ，ベクトル $\mathbf{a}$，定数 $c$ によって

$$ f(\mathbf{x}) = \mathbf{x}^T\mathbf{A}\mathbf{x} + \mathbf{a}^T\mathbf{x}+c = \sum_{i,j}A_{ij}x_ix_j + \sum_ia_ix_i + c$$

と書けるのが二次最適化問題です。これも行列・ベクトルで書けるので線形代数の知識を使うことが出来ます。

### 線型計画問題
$$ \text{minimize}\quad \mathbf{c}^T\mathbf{x}\quad\text{subject to}\quad \mathbf{A}\mathbf{x}=\mathbf{b},\mathbf{x}\geq 0$$
を線型計画問題と呼びます。多くの問題はこの形に帰着するので重要です。
