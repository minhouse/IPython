# 最小二乗法
残差平方和
$$ E = \sum_{n=1}^N |y_n - f(x_n)|^2 $$
が最小となるように$f(x)$を決定する方法。確率論的にはモデルと観測値の誤差が正規分布に従うと仮定した場合の最尤推定法と一致。

```python
N = 100
train_x = random.uniform(0, 1, N)
train_y = random.normal(sin(train_x*pi), 0.1)
plot_x = linspace(0, 1)
scatter(train_x, train_y, s=50, cmap=cm.cool)
```

```python
# 単純な最小二乗法
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(c_[train_x], train_y)
plot(plot_x, reg.predict(c_[plot_x]), color='red')
scatter(train_x, train_y, s=50, cmap=cm.cool)
```

# 基底変換
元々の説明変数$x$を関数$\Psi(x)$で変換した後に線形モデルで回帰を行うことによって、様々な曲線を表現することができる。
例えば
$\Psi(x) = (x^2, x)$
とすると以下のようになる。

```python
def psi(x):
    return array([x**2, x]).T
reg = LinearRegression()
reg.fit(psi(train_x), train_y)
ylim(0, 1.5)
plot(plot_x, reg.predict(psi(plot_x)), color='red')
scatter(train_x, train_y, s=50, cmap=cm.cool)
```

# 正則化
複雑すぎるモデルを利用すると過学習が生じます。これは最小二乗法では単純に学習データに対する精度のみを考えており、学習データ以外は考慮していない為です。

```python
def psi(x):
    return array([x**k for k in range(1, 101)]).T
reg = LinearRegression()
reg.fit(psi(train_x), train_y)
ylim(0, 1.5)
plot(plot_x, reg.predict(psi(plot_x)), color='red')
scatter(train_x, train_y, s=50, cmap=cm.cool)
```

過学習を回避する為には

1. 複雑すぎるモデルは使わない

という方法がありますが、これは現在のマシンパワーに見合わない消極的な方法になります。理想的には過学習を避けつつ出来るだけ表現力の高いモデルを使いたいです。

そこで正則化という考え方が登場します。これは残差平方和の代わりに

$$ E = \sum_{n=1}^N |y_n - f(x_n)|^2 + \alpha (\text{モデルの複雑さに対するペナルティ})$$

を最小化するというものです。$\alpha$はペナルティ項に対する重みです。確率論的にはベイズ確率論における事後確率最大化と同等です。

```python
from sklearn.linear_model import Ridge
def psi(x):
    return array([x**k for k in range(1, 101)]).T
reg = Ridge(alpha=0.01)
reg.fit(psi(train_x), train_y)
ylim(0, 1.5)
plot(plot_x, reg.predict(psi(plot_x)), color='red')
scatter(train_x, train_y, s=50, cmap=cm.cool)
```

# ロジスティック回帰
ロジスティック回帰は回帰と名が付いていますが、主に識別の問題に利用されます。そのモデルは
$$ y = \frac{1}{1 + \exp(w^Tx)} $$
というものです。この関数の値は必ず$0$から$1$の範囲に入ります。つまり、$y$は何らかの確率を表していると考える事ができます。
今、$y$をデータ$x$がクラス$1$に所属する確率を表していると考えると
$$ p(1|x) = \frac{1}{1 + \exp(w^Tx)}$$
を変形してみると
$$ \log \frac{p(1|x)}{1-p(1|x)} = w^Tx $$
となります。つまりオッズの対数が線形モデルであるとするモデルです。

```python
N = 200
train_x = random.uniform(-2, 2, (N, 2))
train_t = train_x[:, 0]**2 < train_x[:, 1]
xlim(-2, 2)
ylim(-2, 2)
scatter(train_x[:, 0], train_x[:, 1], c=train_t, s=50, cmap=cm.cool)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(train_x, train_t)
X, Y = meshgrid(linspace(-2, 2), linspace(-2, 2))
Z = clf.predict(c_[X.ravel(), Y.ravel()]).reshape(X.shape)
xlim(-2, 2)
ylim(-2, 2)
pcolor(X, Y, Z, alpha=0.5)
scatter(train_x[:, 0], train_x[:, 1], c=train_t, s=50, cmap=cm.cool)
```

基底変換として
$$ \Psi(x, y) = (x^2, y^2, xy, x, y)$$
をしてみると以下のようになります。

```python
def psi(x):
    return array([x[:,0]**2, x[:,1]**2, x[:,0]*x[:,1], x[:, 0], x[:, 1]]).T
clf = LogisticRegression()
clf.fit(psi(train_x), train_t)
X, Y = meshgrid(linspace(-2, 2), linspace(-2, 2))
Z = clf.predict(psi(c_[X.ravel(), Y.ravel()])).reshape(X.shape)
xlim(-2, 2)
ylim(-2, 2)
pcolor(X, Y, Z, alpha=0.5)
scatter(train_x[:, 0], train_x[:, 1], c=train_t, s=50, cmap=cm.cool)
```
