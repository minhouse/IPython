板書
授業時の解説内容を随時、私が書いていきます。黒板やホワイトボードへの板書の代わりです。ノート等に手書きをするよりも確実に綺麗で速いと思います。例えば以下のような感じで書いていくと思います。

ベイズの定理
P(A|B)=P(B|A)P(A)P(B)
をベイズの定理(Bayes's theorem)と呼びます。P(B) は A によらない定数なので、もっと簡単に

P(A|B)∝P(B|A)P(A)
と書くことが出来ます。∝というのは左辺は右辺に比例するという意味です。実際の値は確率の総和が 1 になる事をつかって計算します。

例題として有名な「ガン検診」の問題をベイズの定理で解いてみましょう。

ガンであるという事象「ガン」と、ガン検診を受けたら陽性反応が出るという事象「陽性」を考えます。過去の統計から

P(ガン)P(陽性|ガン)P(陽性でない|ガンでない)=0.001=0.99=0.99
である事が分かっているとします。つまりガンである人は1000人に1人の割合で存在し, 診断は99%で正しい結果が出るということです。 この時 P(ガン|陽性) を計算してみましょう。つまり陽性だという結果が出た時に本当にガンである確率です。

ベイズの定理より
P(ガン|陽性)P(ガンでない|陽性)∝P(陽性|ガン)P(ガン)=0.99×0.001∝P(陽性|ガンでない)P(ガンでない)=0.001×0.999
となります。「ガン」と「ガンでない」で全ての場合を網羅しているので足して1になるように右辺を定数倍します。 下の計算結果から
P(ガン|陽性)≈0.09=9%
となります。つまり、この診断は陽性が出てもほとんど信頼できないという事になります。





```python
In [1]:
a = 0.99 * 0.001   # P(ガン|陽性)の右辺
b = 0.01 * 0.999   # P(ガンでない|陽性)の右辺
print a/(a+b)
print b/(a+b)
0.0901639344262
0.909836065574
```

質問・回答
授業後の質問は授業時に作成した文書に直接書き込む形でやるのが良いかと思います。私がそれに回答を書き加えます。例えば以下のような感じで。

[質問] 上の問題で, 診断が99%の正確さなのに結果が9%というのが信じられないのですが、何故このような数字になるのですか？

[回答] ガンの発生率が1000人に1人程度という低頻度なのに、診断は100人に1人くらいの頻度で間違えます。つまりこの例では99%という数字は低すぎるんです。ざっくり計算すると、1000人の人がこの診断を受けると10人くらい陽性という診断を受けます。実際にガンの人は1人程度なので約1/10に近い確率となります。
