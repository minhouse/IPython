# モデルの選択
回帰分析では

$$ y = ax + b $$

などの**モデル(model)**を決めて分析を行いました。回帰分析に限らず、どのような数学的なモデル(模型)を選択するかが非常に重要となります。

## 精度と過剰適合のトレードオフ
以下のデータセットに多項式モデル

$$ y=a_nx^n+a_{n-1}x^{n-1}+\cdots+a_1x+a_0 $$

を当てはめる事を考えます。

```python
N = 10
x = random.uniform(0, 1, N)
y = random.normal(sin(x), 0.1, N)
ylim(0, 1)
scatter(x, y)
```

一次式を当てはめて，残差平方和を計算すると次のようになります。

```python
X = array([x, ones(N)]).T
a = linalg.solve(X.T.dot(X), X.T.dot(y))
pred_x = linspace(0, 1)
pred_y = array([pred_x, ones(len(pred_x))]).T.dot(a)
ylim(0, 1)
scatter(x, y)
plot(pred_x, pred_y)

print u'残差平方和=', sum((y - X.dot(a))**2)
```

2次式を当てはめた場合は次の様になります.

```python
X = array([x**2, x, ones(N)]).T
a = linalg.solve(X.T.dot(X), X.T.dot(y))
pred_x = linspace(0, 1)
pred_y = array([pred_x**2, pred_x, ones(len(pred_x))]).T.dot(a)
scatter(x, y)
plot(pred_x, pred_y)

print (u'残差平方和=', sum((y - X.dot(a))**2))
```

2次式の方が残差平方和が小さくなったと思います. 一般に

> より複雑なモデルの方が学習データに対する精度は高くなる

という事が言えます。より複雑なモデルは曲線の選択肢が増える(探索範囲が増える)ためです.

モデルの複雑さは含まれるパラメータの数に強く依存します. これを**自由度(degree of freedom)**と呼びます. 例えば $y=ax+b$ の自由度は2, $y=ax^2+bx+c$ の自由度は3です。

では次数を上げて10次式にしたらどうなるでしょうか.

```python
X = array([x**k for k in range(11)]).T
a = linalg.solve(X.T.dot(X), X.T.dot(y))
pred_x = linspace(0, 1)
pred_y = array([pred_x**k for k in range(11)]).T.dot(a)
ylim(0, 1)
scatter(x, y)
plot(pred_x, pred_y)

print (u'残差平方和=', sum((y - X.dot(a))**2))
```

更に残差平方和が減りました. しかし見るからにこれは良いモデルではありません。このように学習データの特徴を過剰に学習してしまう事を**過剰適合(over fitting)**と呼びます.

このようにモデルの複雑さの程度には精度と過剰適合のトレードオフが存在するという事がわかります。過剰適合を防ぐ為には3つの対策があります.

1. データ数を大量にする
2. 複雑なモデルを使わない
3. ベイズ的なモデルを使う

1番はもしそれがコスト的に可能ならば、簡単な解決策です. 上の10次式の例も、もしデータが1000個あれば次のように極端な学習は行われません. しかし現実にはデータを集める事が難しい状況が多いです.

```python
N = 1000
x = random.uniform(0, 1, N)
y = random.normal(sin(x), 0.1, N)
X = array([x**k for k in range(11)]).T
a = linalg.solve(X.T.dot(X), X.T.dot(y))
pred_x = linspace(0, 1)
pred_y = array([pred_x**k for k in range(11)]).T.dot(a)
ylim(0, 1)
scatter(x, y)
plot(pred_x, pred_y, color='red')

print (u'残差平方和=', sum((y - X.dot(a))**2))
```

## 汎化性能
モデルの良さを決める指標の１つが**汎化性能(generalization performance)**です.
ある学習データセット $D=\{(\mathbf{x}_1,y_1),\ldots,(\mathbf{x}_n,y_n)\}$ から学習されたモデル

$ y=f(\mathbf{x}) $

の「未知のデータに対する当てはまりの良さ」を汎化性能と呼びます。
汎化性能を最大にするモデルを見つられれば理想ですが、未知のデータに対する精度を厳密に計算する事は不可能です。
しかし、

> 学習データを２つに分け、一方で訓練し他方に対する精度を測る

というやり方で近似的に汎化性能を測る事が出来ます。これを**検証(validation)**と呼びますが，**K交差検証(K-cross validation)**という方法がよく使われます。

----
【K交差検証】

1. 学習データをK組に分ける
2. K-1組を学習し，残り1組に対する精度を測る
3. これを全ての組み合わせについて計算し，精度の平均で汎化性能を評価する

---

上の多項式の例でこれをやってみましょう。K=3とし、scikit-learnのライブラリを使うことにします。

```python
N = 20
x = random.uniform(0, 1, N)
y = random.normal(sin(x), 0.1, N)
ylim(0, 1)
scatter(x, y)
```

まずは2次式モデルの場合

```python
from sklearn.cross_validation import KFold
M = 3

rs = []
for (train_idx, test_idx) in KFold(n = N, n_folds=3):
    train_x = x[train_idx]
    train_y = y[train_idx]
    X = array([train_x**k for k in range(M)]).T
    a = linalg.solve(X.T.dot(X), X.T.dot(train_y))
    pred_x = linspace(0, 1)
    pred_y = array([pred_x**k for k in range(M)]).T.dot(a)
    ylim(0, 1)
    scatter(x, y)
    plot(pred_x, pred_y)
    test_x = x[test_idx]
    test_y = y[test_idx]
    rs.append( sum((test_y - array([test_x**k for k in range(M)]).T.dot(a))**2) )
print (u'残差平方和平均:', average(rs))
```

10次式モデルの場合

```python
M = 10

rs = []
for (train_idx, test_idx) in KFold(n = N, n_folds=3):
    train_x = x[train_idx]
    train_y = y[train_idx]
    X = array([train_x**k for k in range(M)]).T
    a = linalg.solve(X.T.dot(X), X.T.dot(train_y))
    pred_x = linspace(0, 1)
    pred_y = array([pred_x**k for k in range(M)]).T.dot(a)
    ylim(0, 1)
    scatter(x, y)
    plot(pred_x, pred_y)
    test_x = x[test_idx]
    test_y = y[test_idx]
    rs.append( sum((test_y - array([test_x**k for k in range(M)]).T.dot(a))**2) )
print (u'残差平方和平均:', average(rs))
```

10次モデルの方では残差平方和の平均が大きくなってしまっている事から，汎化性能が低下してしまっている事がわかります。
