次のコードはグラフを書くためのものなので気にしないで下さい。

```python
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d
class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)
```

微分学の基本的な定理である

> $f(x)$が微分可能な関数ならば，$f(x)$が最小値を取る点で $f'(x) = 0$となる

という事を利用して、「$f(x)$ の最小値を求める問題」を「$f'(x) = 0$」を解く問題に帰着させる事が出来ます。
上の逆は成立しないので「$f'(x) = 0$」を解いても、それが最小値を与えるとは限りません。
最大値かもしれませんし、極値かもしれませんし、極値ですらないかもしれません。

従って、十分に注意して利用しなければならないという欠点がありますが、非常に高速であるという特徴があります。

# ニュートン法
ニュートン法とは**反復法**の１つで、$f(x) = 0$ という形の方程式を解くためのものです。まず1次元の場合を考えます。ニュートン法では$f(x) = 0$を解くために、

$$\color{red}{ x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)} }$$

という反復式を使います。これは図形的には $x=x_n$ における $y=f(x)$ の接線 $y=f'(x_n)(x-x_n) + f(x_n)$ と $x$ 軸の交点を $x_{n+1}$ とするものです。下図のように、$x_{n+1}$ は $x_n$ よりも $f(x)=0$の解に近づきます。

```python
from IPython.display import Image
Image("image/newton-method.png")
```

ここで
$$f'(x) = 2x$$
なので、
$$x_{n+1}=x_n - \frac{x_n^2-2}{2x_n} \left( = \frac{x_n^2 + 2}{2x_n}\right )$$

## ニュートン法の特徴
ニュートン法は**二次収束**という非常に速い速度で収束します。(勾配法は1次収束で、ニュートン法は圧倒的に高速です)
欠点は、一回の反復あたりの計算量が大きい(複雑な式を計算する為)事、微分計算が難しい場合が存在する事などです。

### 例
先ほどの $f(x)=x^2-2$ を解く例を**最急降下法**とニュートン法で比較してみます。どちらも初期値は $x_0 = 5.0$ とし $|x_{n+1}-x_n|< 10^{-10}$ を終了条件として比較します。

```python
x0 = 5.0
EPS = 1.0e-10
# 再急降下法
x = x0
for k in range(1000):
    print (x)
    x_new = x - 0.1 * 2*x   # alpha=0.1としてみた
    if abs(x_new - x) < EPS: break
    x = x_new
print ("count=",k)

#ニュートン法
x = x0
for k in range(1000):
    print (x)
    x_new = (x**2+2)/(2*x)
    if abs(x_new - x) < EPS: break
    x = x_new
print ("count=",k)
```

再急降下法では約100回も反復が必要ですが、ニュートン法は数回で終了しています。

## ニュートン法の理論
授業時には詳しく説明しませんでしたが、ニュートン法の理論を書いておきます。興味があったら読んで下さい。

$$ g(x) = x - \frac{f(x)}{f'(x)}$$
とおきます。 $x_{n+1} = g(x_n)$ が成り立ちます。
ここで， $f(x)=0$ の解を $\alpha$ として $x=\alpha$ の周りで $g(x)$ を1次の項までテイラー展開すると

$$ g(x) = g(\alpha) + g'(\alpha)(x-\alpha) + \frac{g''(c)}{2}(x-\alpha)^2 $$

となります。ここで
$$ g(\alpha) = \alpha\qquad(\because f(\alpha)=0)$$
$$ g'(\alpha) = 0 \qquad \left(\because g'(x) = \frac{f(x)f''(x)}{\{f'(x)\}^2}\right)$$
なので

$$ g(x) = \alpha + \frac{g''(c)}{2}(x-\alpha)^2 $$

です。これに $x=x_n$ を代入すると

$$ x_{n+1} = \alpha + \frac{g''(c)}{2}(x_n-\alpha)^2 $$

となるので

$$|x_{n+1}-\alpha| = \left|\frac{g''(c)}{2}\right||x_n-\alpha|^2 $$

が成り立ちます。つまり $x_{n+1}$ の誤差が $x_n$ の**誤差の二乗**にほぼ比例するという事になります。これがニュートン法が高速である理由です。

# ニュートン法を使った最小値計算

冒頭で説明したように

> $f(x)$ の最小値を求める事 $\Rightarrow$ $f'(x) = 0$を解くこと

という読み替えを使います。ニュートン法の公式を $f'(x) = 0$ に当てはめると

$$\color{red}{ x_{n+1} = x_n - \frac{f'(x_n)}{f''(x_n)} }$$

という反復式になります。

## 勾配法との関係
勾配法は(一変数の場合)

$$x_{n+1} = x_n - \alpha f'(x_n) $$

という更新式を使います。これをニュートン法を見比べると

$$ \alpha \rightarrow \frac{1}{f''(x_n)} $$

となっています。$f''(x)$ の大きさは曲線の曲がり具合を表していて、直線ならば $0$, カーブが急であるほど大きな値を取ります。
つまり、ニュートン法の更新式は

- 曲線の曲がり具合が小さいときは大きな歩幅で進み
- 曲がり具合が大きい時には小さな歩幅で進む

というように、歩幅を適応的に変化させる手法と見なす事が出来ます。

###例
$$ f(x) = \frac{x}{\log x}$$
の最小値を求めてみたいと思います。

$$ f'(x) = \frac{\log x - 1}{(\log x)^2}$$
$$ f''(x) = \frac{2-\log x}{x (\log x)^3}$$

すると更新式は

$$ x_{n+1} = x_n - \frac{f'(x_n)}{f''(x_n)} = x_n - \frac{\log x-1}{(\log x)^2}\frac{x(\log x)^3}{2-\log x} = \frac{x_n\{(\log x_n)^2-2\}}{\log x_n-2}$$

となります。

```python
from math import log
x = 3.0
for k in range(10):
    print (x)
    x = x*((log(x))**2 - 2)/(log(x)-2)
```

グラフを描いてみると下図のようになっていて、確かに $x = 2.7$付近で最小値を取っています。(実はこの値は自然対数の底 $e=2.718281828\cdots$です。)

```python
x = linspace(2.0, 3.5)
plot(x, x/log(x))
```

## ニュートン法(高次の場合)

一変数の場合は

$$x_{n+1} = x_n - \frac{f'(x_n)}{f''(x_n)} $$

という漸化式でしたが、次はこれの多変数版を紹介します。まず、右辺分子の一階導関数は勾配$\nabla f(x_n)$ に置き換えられます。続いて、右辺分母の二階導関数ですが、これは**ヘッセ行列(Hessian matrix)**という行列に一般化されます。関数 $f$ に対するヘッセ行列は

$$ \nabla^2 f(\mathbf{x}) \stackrel{\mathrm{def}}{=} \begin{pmatrix}
\frac{\partial^2 f}{\partial x_1\partial x_1} & \frac{\partial^2 f}{\partial x_1\partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_1\partial x_n} \\
\frac{\partial^2 f}{\partial x_2\partial x_1} & \frac{\partial^2 f}{\partial x_2\partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_2\partial x_n} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial^2 f}{\partial x_n\partial x_1} & \frac{\partial^2 f}{\partial x_n\partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_n\partial x_n} \\
\end{pmatrix}$$
によって定義されます。これが分母に来るので逆行列を取って

$$ \color{red}{ \mathbf{x}_{n+1} = \mathbf{x}_n - (\nabla^2f(\mathbf{x}_n))^{-1}\nabla f(\mathbf{x}_n) } $$

となります。これが多次元版のニュートン法です。この公式の厳密な導出はここでは省略します。
今, $f$ として十分滑らかな関数を考えると、勾配が $\mathbf{0}$ になる点でヘッセ行列が正定値ならば $f$ は極小値を取ります。またヘッセ行列は対称行列になる事にも注意しましょう。従って、以下ではヘッセ行列が正定値対称行列である場合のみを考えます。

### 例
$$ f(x,y) = x^2y^2+4x^2+y^2$$
に対してニュートン法を利用してみましょう。まず勾配とヘッセ行列を計算してみると以下のようになります。

$$\nabla f= (2xy^2+8x, 2x^2y+2y)$$
$$\nabla^2 f = \begin{pmatrix}
2y^2+8 & 4xy \\
4xy    & 2x^2+2
\end{pmatrix}$$

以下は初期値を $(1,1)$ として走らせた例です。

```python
EPS = 1.0e-5
x = array([1.0, 1.0])
for k in range(1000):
    print (x)
    diff = array([2*x[0]*x[1]**2 + 8*x[0], 2*x[0]**2*x[1] + 2*x[1]]) # 勾配
    hess = array([[2*x[1]**2 + 8, 4*x[0]*x[1]], [4*x[0]*x[1], 2*x[0]**2 + 2]]) # ヘッセ行列
    new_x = x - linalg.inv(hess).dot(diff) # numpy.linalg.invで逆行列を求める
    if linalg.norm(new_x - x) < EPS: break
    x = new_x
print ("count=",k+1)
```

するとたった3ステップで $(0,0)$ に収束しました。反復の様子を図示してみると次のようになります。

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X, Y = meshgrid(linspace(-1, 1, 20), linspace(-1, 1, 20))
xlim(-1, 1)
ylim(-1, 1)
Z = X**2*Y**2 + 4*X**2 + Y**2
ax.plot_wireframe(X, Y, Z)
EPS = 1.0e-5
x = array([1.0, 1.0])
traj = []
for k in range(1000):
    traj.append([x[0], x[1], x[0]**2*x[1]**2+4*x[0]**2+x[1]**2])
    diff = array([2*x[0]*x[1]**2 + 8*x[0], 2*x[0]**2*x[1] + 2*x[1]]) # 勾配
    hess = array([[2*x[1]**2 + 8, 4*x[0]*x[1]], [4*x[0]*x[1], 2*x[0]**2 + 2]]) # ヘッセ行列
    new_x = x - linalg.inv(hess).dot(diff) # numpy.linalg.invで逆行列を求める
    if linalg.norm(new_x - x) < EPS: break
    x = new_x
traj = array(traj)
scatter(traj[:,0], traj[:,1], zs=traj[:,2], s=30, c="red")
for k in range(1,len(traj)):
    v = array([traj[k-1,:], traj[k,:]]).T
    ax.add_artist(Arrow3D(v[0], v[1], v[2], mutation_scale=10,arrowstyle="-|>", color="k"))
ax.view_init(elev=60)
```

### 例
最後の例として二次関数
$$f(\mathbf{x}) = \frac{1}{2}||\mathbf{A}\mathbf{x}-\mathbf{b}||^2$$
$$ \mathbf{A} = \begin{pmatrix} 1 & 2 & 3 \\ 2 & 3 & 1 \\ 1 & 3 & 2 \end{pmatrix},\quad \mathbf{b} = \begin{pmatrix} 2 \\ 1 \\ 1 \end{pmatrix} $$
の最小化問題にニュートン法を用いてみます。勾配法で同じ問題を解いた際には収束までに400ステップ以上必要でした。
まず、勾配とヘッセ行列を計算してみると以下のようになります。

$$\nabla f(\mathbf{x}) = \mathbf{A}^T(\mathbf{A}\mathbf{x}-\mathbf{b}) $$
$$\nabla^2 f(\mathbf{x}) = \mathbf{A}^T\mathbf{A} $$

従ってニュートン法の更新式は

$$\mathbf{x}_{n+1} = \mathbf{x}_{n} - (\mathbf{A}^T\mathbf{A})^{-1}\mathbf{A}^T(\mathbf{A}\mathbf{x}_n-\mathbf{b})$$

となりますが，

$$(\mathbf{A}^T\mathbf{A})^{-1}\mathbf{A}^T\mathbf{A} = \mathbf{I}\quad(\text{単位行列}) $$

なので

$$\mathbf{x}_{n+1} = \mathbf{x}_n - (\mathbf{x}_n-(\mathbf{A}^T\mathbf{A})^{-1}\mathbf{A}^T\mathbf{b}) = (\mathbf{A}^T\mathbf{A})^{-1}\mathbf{A}^T\mathbf{b} $$ 

となります。なんと、$\mathbf{x}_n$ が消えてしまいましたが、これは1ステップで正しい解に辿り着けるという事を意味します。計算してみましょう。

```python
A = array([[1,2,3],[2,3,1],[1,3,2]], dtype=float)
b = array([2,1,1], dtype=float)
x = linalg.inv(A.T.dot(A)).dot(A.T.dot(b)) # linalg.invで逆行列を計算できる。
print (x)
```

ちなみに勾配法で400ステップ掛けて計算した解は `x= [ 0.66613273 -0.33299493  0.6665689 ]` でした。ニュートン法では一ステップでより高い精度を実現しています。

## 準ニュートン法

これまで見たようにニュートン法は極めて収束が速いという特徴があるのですが、勾配法に比べて使いにくい点を列挙すると以下のようになります。

1. 極小値に収束する保証がない。極大値や停留点にも収束したり、発散したりする。
2. ヘッセ行列及びその逆行列の計算が必要。

まず1についてですが、ヘッセ行列 $\nabla^2 f$ は一般に正定値では無い為にこういった問題が生じます。極小値の周辺で $\nabla^2 f$ が正定値である範囲内に初期値を取らなければならないので、収束域が非常に狭くなってしまいます。

2についてですが、一般に $n\times n$ 行列の逆行列の計算には $\mathcal{O}(n^3)$ の計算時間が必要になります。つまり、収束までの反復回数は減りますが、一回の更新に必要な計算量が増加してしまうので、このままでは逆に総計算量が増加してしまう場合もあります。

**準ニュートン法 (quasi newton method)**という手法はこれらの問題を解決するもので、特にBFGS公式と呼ばれるものは非線形最適化問題の解法のうちで現在最も汎用的で効率が良いアルゴリズムであると言われます。次の節ではこの詳しい説明をしますが、難しいかもしれないので今はスキップしても良いです。

## ヘッセ行列の近似

これまで何度か登場した**正定値(positive definite)**という概念について復習すると、$n\times n$行列 $\mathbf{A}$ が正定値であるとは任意の$n$次行列 $\mathbf{x}$ に対して$ \mathbf{x}^T\mathbf{A}\mathbf{x} > 0$が成り立つ事です。

ニュートン法の更新式は
$$ \mathbf{x}_{n+1} = \mathbf{x}_n - (\nabla^2 f(\mathbf{x}_n))^{-1}\nabla f(\mathbf{x}_n)$$
でしたが，**ヘッセ行列 $\nabla^2 f(\mathbf{x}_n)$ が正定値ならば進行方向は下降方向である**という事が言えます。
簡単に説明すると、正定値対称行列の逆行列も正定値なので，
$$ - \nabla f(\mathbf{x}_n)^T(\nabla^2 f(\mathbf{x}_n))^{-1}\nabla f(\mathbf{x}_n) < 0 $$
が成り立ちます。つまりベクトル$\nabla f(\mathbf{x}_n)$(上昇方向)とベクトル$-(\nabla^2 f(\mathbf{x}_n))^{-1}\nabla f(\mathbf{x}_n)$のなす角が$90^\circ$以上になるからです。

ヘッセ行列が正定値でない場合には、値が増加していって極大値にたどり着いてしまったり無限大に発散してしまうなどの問題が生じます。そこで、準ニュートン法ではヘッセ行列 $\nabla^2 f(\mathbf{x}_n)$ やその逆行列を、正定値対称行列 $\mathbf{B}_n,\mathbf{H}_n$ で近似する事によってこの問題を回避します。

$$\mathbf{B}_{n} \approx \nabla^2 f(\mathbf{x}_n),\qquad\mathbf{H}_{n} \approx (\nabla^2 f(\mathbf{x}_n))^{-1}$$

この時 $\mathbf{B}_n$ が満たすべき条件として**セカント条件(secant equation)**と呼ばれる

$$ \nabla f(\mathbf{x}_{n+1}) - \nabla f(\mathbf{x}_n) = \mathbf{B}_{n+1}(\mathbf{x}_{n+1}-\mathbf{x}_n)\quad\text{もしくは}\quad
 \mathbf{H}_{n+1}\left(\nabla f(\mathbf{x}_{n+1}) - \nabla f(\mathbf{x}_n)\right) = \mathbf{x}_{n+1}-\mathbf{x}_n$$

を考えます。一変数の場合、この式は

$$ \frac{f'(x_{n+1})-f'(x_n)}{x_{n+1}-x_n} = b_{n+1} \approx f''(x_{n+1})$$

と書けるのでイメージできると思います。

## 直線探索

$\mathbf{x}_{n+1}$を適当に選んでしまっては、セカント条件
$$ \nabla f(\mathbf{x}_{n+1}) - \nabla f(\mathbf{x}_n) = \mathbf{B}_{n+1}(\mathbf{x}_{n+1}-\mathbf{x}_n)$$
を満たす正定値対称行列 $\mathbf{B}_{n+1}$ が存在する保証がありません。また単に下降方向に進むだけでは、収束を保証する事は出来ません。$\mathbf{x}_{n+1}$ を上手く選ぶ必要があります。これを以下のようにステップサイズ $\alpha$ を上手く選ぶ問題として考えます。

1. $\mathbf{x}_0$ を適当に決める。
2. $\mathbf{H}_0 = \mathbf{I}$とする。(注:単位行列は正定値対称行列)
3. $n=0,1,\ldots$について収束するまで以下を繰り返す。
   1. 進行方向を $\mathbf{p}_n = -\mathbf{H}_n\nabla f(\mathbf{x}_n)$ とする。($\mathbf{H}_n$ が正定値対称なのでこれは下降方向)
   2. **ステップサイズ($\alpha$)を適切に選んで** $\mathbf{x}_{n+1} = \mathbf{x}_n + \alpha\mathbf{p}_n$ とする。
   3. $\mathbf{H}_{n+1}(\nabla f(\mathbf{x}_{n+1})-\nabla f(\mathbf{x}_n))=\mathbf{x}_{n+1}-\mathbf{x}_n$ を満たす $\mathbf{H}_{n+1}$ を求める。

まず、どのような $\alpha > 0$ が適切であるかについてですが、以下の**Armijo,Wolfeの条件**を満たすならば大域的な収束性(どのような初期値から出発してもいずれかの極小値に収束する)が保証されます。

----
【Armijo, Wolfeの条件】

$0 < c_1 < c_2 < 1$ として
$$f(\mathbf{x}_n+\alpha\mathbf{p}_n)\leq f(\mathbf{x}_n) + c_1\mathbf{p}_n^T\nabla f(\mathbf{x}_n)\qquad \text{(Armijoの条件)}$$
$$ \nabla f(\mathbf{x}_n+\alpha\mathbf{p}_n)^T\mathbf{p}_n \geq c_2\nabla f(\mathbf{x}_n)^T\mathbf{p}_n \qquad \text{(Wolfeの条件)}$$

を満たすような $\alpha$ であれば良い。よく用いられるパラメータの値は $c_1 = 10^{-4}, c_2 = 0.9$。

----

直線探索は例えば以下の手順で行います。

----
【直線探索】

$0 < \gamma < 1, 1 < \delta$ を $\gamma\delta > 1$ を満たすように取る。
1. $\alpha = 1$ とする
2. Armijoの条件を満たさないならば $\alpha$ は大きすぎるので $\alpha \leftarrow \gamma\alpha$ とし2に戻る。
3. Wolfeの条件を満たさないならば $\alpha$ は小さすぎるので $\alpha \leftarrow \delta\alpha$ とし2に戻る。
4. 両方の条件を満たしたら終了。

----

## BFGS公式
Armijo,Wolfe条件を満たす $\alpha$ を求めれば、セカント条件

$$\mathbf{H}_{n+1}(\nabla f(\mathbf{x}_{n+1}) - \nabla f(\mathbf{x}_n)) = \mathbf{x}_{n+1}-\mathbf{x}_n$$

を満たす $\mathbf{H}_{n+1}$ を求める事が可能になります。以下のBFGS公式は現在最も良く用いられるものです。

----
【BFGS公式】

$$ \mathbf{H}_{n+1} = \mathbf{H}_n + \frac{\gamma_n + \mathbf{y}_n^T\mathbf{H}_n\mathbf{y}_n}{\gamma_n^2}\mathbf{s}_n\mathbf{s}_n^T-\frac{1}{\gamma_n}\left(\mathbf{H}_n\mathbf{y}_n\mathbf{s}_n^T+(\mathbf{H}_n\mathbf{y}_n\mathbf{s}_n^T)^T\right) $$
但し， $\mathbf{s}_n = \mathbf{x}_{n+1}-\mathbf{x}_n, \mathbf{y}_n=\nabla f(\mathbf{x}_{n+1})-\nabla f(\mathbf{x}_n), \gamma_n = \mathbf{s}_n^T\mathbf{y}_n$。

----

この公式では前のステップでの $\mathbf{H}_n$ から次のステップでの $\mathbf{H}_{n+1}$ を計算しています。この更新の計算量は $\mathcal{O}(n^2)$ です。直接ヘッセ行列の逆行列を計算する為には $\mathcal{O}(n^3)$ の計算量が必要なので大変効率の良いものとなっています。また、勾配 $\nabla f$ の計算が出来れば十分なので二階微分の計算も不要です。

## 例
では、BFGS法を実装してみましょう。ここまで複雑になってくると幾つかの関数に分けて実装します。

```python
#### BFGS公式 による準ニュートン法

# 直線探索
# f: 対象関数, grad: 勾配関数, x0: 現在位置, p0: 進行方向, grad0=grad(x0)
def linesearch(f, grad, x0, p0, grad0):
    C1 = 1.0e-4
    C2 = 0.9
    GAMMA = 0.9
    DELTA = 1.2
    ARMIJO_RHS = f(x0) + C1*p0.dot(grad0) # Armijo条件の右辺
    WOLFE_RHS  = C2*grad(x0).dot(p0)      # Wolfe条件の右辺
    ALPHA = 1.0
    for k in range(1000):
        x1 = x0 + ALPHA*p0
        if f(x1) > ARMIJO_RHS:
            # Armijo条件を満たさない
            ALPHA *= GAMMA
            continue
        if grad(x1).dot(p0) < WOLFE_RHS:
            # Wolfe条件を満たさない
            ALPHA *= DELTA
            continue
        return ALPHA
    raise Exception(u"Failed to find acceptable step size")
    
# BFGS公式 (ヘッセ行列の逆行列の近似行列Hを更新)
# H: 現在のH (上書きする), x0, x1: 現在位置と次の位置, grad0, grad1: 現在の勾配と次の勾配
def updateH(H, x0, x1, grad0, grad1):
    s = x1 - x0
    y = grad1 - grad0
    gamma = s.dot(y)
    A = outer(s, s)
    B = outer(H.dot(y), s)
    c = y.dot(H.dot(y))
    H += (gamma + c)/gamma**2*A - (B + B.T)/gamma

# f: 対象関数, grad: 勾配関数, x: 初期値
def quasi_newton(f, grad, x0):
    EPS = 1.0e-5
    N   = len(x0)     # 次元
    H   = identity(N) # Hは単位行列で初期化
    grad0 = grad(x0)
    for k in range(1000):
        p     = -H.dot(grad0)               # 進行方向
        ALPHA = linesearch(f, grad, x0, p, grad0) # 直線探索
        x1    = x0 + ALPHA*p                # 次の位置
        if linalg.norm(x1 - x0) < EPS:
            return x1, k+1   # 解と反復回数を返す
        grad1 = grad(x1)
        updateH(H, x0, x1, grad0, grad1) # Hを更新
        x0 = x1
        grad0 = grad1
    raise Exception("quasi_newton did not converge")
```

先ほどやったローゼンブロック関数にこれを使ってみましょう。

```python
def f(x):
    return (1-x[0])**2 + 100*(x[1]-x[0]**2)**2
def grad(x):
    return array([-2*(1-x[0])-400*x[0]*(x[1]-x[0]**2), 200
                  *(x[1]-x[0]**2)])
print (quasi_newton(f, grad, array([-1.0, 2.0])))
```

約50ステップで収束しました。勾配法では約2000ステップでしたから大分高速です。近似を行っている分ニュートン法よりは反復回数が多いですが、１ステップあたりの更新が高速であるのと、収束性が保証されている事が重要です。

簡単な例として $f(x)=\sin x$ の極小値をニュートン法で求める事を考えてみましょう。初期値を $x=1.0$ にすると以下のように $\pi/2 = 1.57\cdots$ に収束しますが、これは極大値です。

```python
x = 1.0
for k in range(1000):
    new_x = x - cos(x)/(-sin(x))
    if abs(new_x - x) < EPS: break
    x = new_x
print (x)
```
