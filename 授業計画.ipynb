
1. 最適化問題の解法(全5～6時間)
    * 目標: 制約付き非線型最適化問題を解けるようになる。
    * 最急勾配法・共役勾配法・準ニュートン法・確率的勾配降下法・BFGS法など
2. ランダムサンプリング
    * 目標: MCMCを実装出来るようになる(全6～7時間)
    * 確率分布の変数変換・逆変換法・インポータンスサンプリング・棄却サンプリング・マルコフ連鎖モンテカルロ法(MCMC法) 
3. 古典統計学(全6～7時間)
    * 目標: 代表的な分布の理解、仮説検定の理解
    * 正規分布・ポアソン分布・二項分布・ガンマ分布・t分布・... どのような場面でどの分布が用いられるのかといった内容の理解.
4. ベイズ統計学(全5～6時間)
    * 目標: ベイズ推定の理解
    * 確率のベイズ的解釈・ベイズの定理・MAP推定・予測分布の計算など.
