```python
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d
class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)
```

# 勾配法とは

最小化問題を厳密に解くのは一般には非常に難しいです。例えば，$f(x)=x^2$ が最小になるのは $x=0$ の時と簡単に判りますが、一般の関数ではそうは行きません。
**勾配法(gradient method)**は**非線形最適化問題**を解くために利用される標準的な方法です。この方法は非常にシンプルなので理解が容易であり、様々な拡張によって様々な問題に適用する事が出来ます。

勾配法は**反復法(iterative method)**の一種です。反復法というのは

1. 初期値 $x_0$ を定める。
2. 何らかのルールに従って $x_n$ から $x_{n+1}$ を計算するという事を繰り返す
3. $x_n$ が収束したらそれを答えとする

という形のアルゴリズムです。

勾配法の考え方は非常にシンプルで、関数の**勾配(gradient)**を見て値が減少する方向に下っていくというものです。下図のようなイメージです。

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X, Y = meshgrid(linspace(-1, 1, 20), linspace(-1, 1, 20))
xlim(-1, 1)
ylim(-1, 1)
Z = X**2+2*Y**2
ax.plot_wireframe(X, Y, Z)
# 最急降下法(後で説明)
x = np.array([1.0, 1.0])
ALPHA = 0.1
traj = np.zeros((20,3))
for k in range(20):
    traj[k,:] = [x[0], x[1], x[0]**2 + 2*x[1]**2]
    x -= ALPHA*2*np.array([x[0],2*x[1]])
scatter(traj[:,0], traj[:,1], zs=traj[:,2], s=30, c="red")
for k in range(1,20):
    v = np.array([traj[k-1,:], traj[k,:]]).T
    ax.add_artist(Arrow3D(v[0], v[1], v[2], mutation_scale=10,arrowstyle="-|>", color="k"))
ax.view_init(elev=60)
```

### 例

$$ f(x,y) = x^2 + 2y^2 $$
という関数の勾配は

$$\nabla f(x,y) = (2x, 4y) $$
となります。例えば点 $(x,y)=(1,1)$ の勾配は $\nabla f(1,1) = (2,4)^T$ となるので $\mathbf{v}=(1,2)$ の方向が最も増加率が大きいです。さらに増加率の大きさは $||\nabla f(1,1) = \sqrt{2^2+4^2} = 2\sqrt{5}$ です。実際に見てみると下図のようになっています。

勾配法(にかかわらず多くの反復解法)では、厳密に最小な解を求める事が出来るとは限りません。また、反復が収束する保証もありません。このことは以下の具体例で確認してみます。

## 勾配

多変数関数 $f(\mathbf{x})$ について、これを $x_1,x_2,\ldots,x_n$ で偏微分したものからなるベクトルを $f$ の**勾配**と呼び、$\nabla f(\mathbf{x})$と書きます。

つまり

$$\nabla f(\mathbf{x}) \overset{\text{def}}{=} \left(\frac{\partial f}{\partial x_1}, \ldots, \frac{\partial f}{\partial x_n}\right) $$

と定義されます。$\mathbf{x}=\mathbf{a}$ における勾配 $\nabla f(\mathbf{a})$ は、$f(\mathbf{x})$ の増加率が最も大きい(つまり、傾きが最も急な増加方向)を向いていて、その大きさ $||\nabla f(\mathbf{a})||$ が増加率の大きさを表します。

### 例

$$ f(x,y) = x^2 + 2y^2 $$
という関数の勾配は

$$\nabla f(x,y) = (2x, 4y) $$
となります。例えば点 $(x,y)=(1,1)$ の勾配は $\nabla f(1,1) = (2,4)^T$ となるので $\mathbf{v}=(1,2)$ の方向が最も増加率が大きいです。さらに増加率の大きさは $||\nabla f(1,1) = \sqrt{2^2+4^2} = 2\sqrt{5}$ です。実際に見てみると下図のようになっています。

```python
fig = figure()
ax = Axes3D(fig)
X, Y = meshgrid(linspace(0, 5, 20), linspace(0, 5, 20))
Z = X**2+2*Y**2
ax.plot_wireframe(X, Y, Z)
scatter([1],[1],zs=[3],s=30,c="red")
ax.add_artist(Arrow3D([1, 1+1], [1, 1+2], [3, 3], mutation_scale=20,arrowstyle="-|>"))
ax.contour3D(X, Y, Z, 20)
```

等高線とベクトル v=(1,2) を描いてみると下図のようになります。等高線とベクトルが直交していることから、最も急な方向を向いている事が分かります。

```python
contour(X, Y, Z)
arrow(1, 1, 1, 2, head_width=0.1)
axes().set_aspect('equal')
```

## 最急降下法

**再急降下法(steepest descent method)**は最も単純な勾配法です。アイデアは非常にシンプルで「最も減少率が大きい方向に下る」というものです。$\nabla f(\mathbf{x})$ が最も増加率の大きい方向でしたから

$$ (\text{最も減少率の大きい方向}) = -\nabla f(\mathbf{x})\text{の方向} $$

です。従って再急降下法の更新式は

$$ \color{red}{\mathbf{x}_{n+1} = \mathbf{x}_n - \alpha \nabla f(\mathbf{x}_n) }$$

となります。ただし $\alpha > 0$ は一歩の歩幅(ステップサイズ)を調整するパラメータです。

## 例
$f(x) = x\log x$ の最小値を求めてみます。一変数関数なので勾配は傾き $f'(x)$ と一致します。従って更新式は

$$ x_{n+1} = x_n - \alpha f'(x_n) = x_n - \alpha (\log x_n + 1) $$

となります。

```python
from math import log
ALPHA = 0.5
EPS   = 1.0e-5
x = 1.0
for k in range(1000):  # 無限ループを避けるために最大反復回数を必ず設定する。
    print (x)
    x_new = x - ALPHA*(log(x) + 1)
    if abs(x_new - x) < EPS: # x(n) と x(n+1) の差が十分小さくなったら終了
        break
    x = x_new
```

x=0.37 あたりに収束しています。実際にグラフを描いてみると次のようになっています。

```python
x = linspace(0, 1)
plot(x, x*np.log(x))
```

$\alpha$ のセッティングには注意が必要です。 $\alpha$ が小さすぎると反復回数が増えてしまいます。

```python
ALPHA = 0.1
EPS   = 1.0e-5
x = 1.0
for k in range(1000):  # 無限ループを避けるために最大反復回数を必ず設定する。
    print (x)
    x_new = x - ALPHA*(log(x) + 1)
    if abs(x_new - x) < EPS: # x(n) と x(n+1) の差が十分小さくなったら終了
        break
    x = x_new
```


α が大きすぎると今度は振動するなどして収束しなくなってしまいます。

```python
ALPHA = 0.8
EPS   = 1.0e-5
x = 1.0
for k in range(1000):
    print (x)
    x_new = x - ALPHA*(log(x) + 1)
    if abs(x_new - x) < EPS:
        break
    x = x_new
```

このように α=0.8では 0.18 あたりと0.75 あたりを行ったり来たりして収束しません。一歩の幅が大きすぎるので、最小値をまたいで振動がおきてしまっています。

```python
ALPHA = 0.8
EPS   = 1.0e-5
x = 1.0
xs = [x]
for k in range(1000):
    x = x - ALPHA*(math.log(x) + 1)
    xs.append(x)

x=linspace(0, 1)
plot(x, x*np.log(x))
for k in range(1, min(len(xs), 10)):
    p0 = [xs[k-1], xs[k-1]*log(xs[k-1])]
    p1 = [xs[k], xs[k]*log(xs[k])]
    plot([p0[0], p1[0]], [p0[1], p1[1]], c="k")
    if k < 10: annotate(str(k), xy=p0, xytext=(p0[0], p0[1]-0.05), arrowprops=dict(arrowstyle="->"))
```

となるので，更新式は

$$ (x_{n+1}, y_{n+1}) = (x_n, y_n) - \alpha\left(-\frac{2x_n\cos y_n}{(x_n^2+1)^2}, -\frac{\sin y_n}{x_n^2+1}\right)$$

## ステップサイズの計算

これまでは $\alpha$ を定数として計算してきましたが、動的に $\alpha$ の値を変化させながら反復するという改良を考える事が出来ます。幾つか考え方がありますが，$\mathbf{p}_n$が進行方向を表しているとして

> $f(\mathbf{x}_{n+1}) = f(\mathbf{x}_n + \alpha\mathbf{p}_n)$ が最小となるような $\alpha$ を選ぶ

という方針を取ります。

## 二次計画問題の場合
最適な$\alpha$を厳密に求める事は一般には困難ですが、二次計画問題では解析的に解くことが出来ます。

二次計画問題の目的関数は$\mathbf{A}$ を**正定値対称行列(positive definite symmetric matrix)**、$\mathbf{b}$ をベクトル， $c$ を定数として
$$f(\mathbf{x}) = \frac{1}{2}\mathbf{x}^T\mathbf{A}\mathbf{x} - \mathbf{b}^T\mathbf{x} + c $$
と書けます。 $1/2$ は微分計算をシンプルにする為に外に出しています。
今，$\mathbf{x}_n$ から $\mathbf{p}_n$ の方向に進むならば $\mathbf{x}_{n+1} = \mathbf{x}_n + \alpha\mathbf{p}_n$ なので
$$ f(\mathbf{x}_{n+1}) = \frac{1}{2}(\mathbf{x}_n+\alpha\mathbf{p}_n)^T\mathbf{A}(\mathbf{x}_n+\alpha\mathbf{p}_n)-\mathbf{b}^T(\mathbf{x}_n+\alpha\mathbf{p}_n)+c $$
となります。これが最小となるような $\alpha$ を探したいのでこれを $\alpha$ で微分します。
$$ \frac{\mathrm{d}}{\mathrm{d}\alpha}f(\mathbf{x}_{n+1}) = \alpha\mathbf{p}_n^T\mathbf{A}\mathbf{p}_n+\mathbf{p}_n^T\mathbf{A}\mathbf{x}_n-\mathbf{b}^T\mathbf{p}_n = \alpha\mathbf{p}_n^T\mathbf{A}\mathbf{p}_n+\mathbf{p}_n^T(\mathbf{A}\mathbf{x}_n-\mathbf{b})$$
$f(\mathbf{x}_{n+1})$ は$\alpha$の二次式なので、これが $0$ になる所で最小になりますから
$$\alpha = - \frac{\mathbf{p}_n^T(\mathbf{A}\mathbf{x}_n-\mathbf{b})}{\mathbf{p}_n^T\mathbf{A}\mathbf{p}_n}$$
となります。この結果は重要なのでまとめておきましょう。

----

【命題2.1】

$$f(\mathbf{x}) = \frac{1}{2}\mathbf{x}^T\mathbf{A}\mathbf{x} - \mathbf{b}^T\mathbf{x} + c $$
の最小化問題において，$\mathbf{x}_n$から $\mathbf{p}_n$ 方向に進む際の最適なステップ幅パラメータ $\alpha$ は
$$\alpha = - \frac{\mathbf{p}_n^T(\mathbf{A}\mathbf{x}_n-\mathbf{b})}{\mathbf{p}_n^T\mathbf{A}\mathbf{p}_n}$$
である。特に，最小勾配法の場合は $\mathbf{p}_n = -\nabla f(\mathbf{x}_n) = -(\mathbf{A}\mathbf{x}_n-\mathbf{b})$ であるから
$$\alpha = \frac{\mathbf{r}_n^T\mathbf{r}_n}{\mathbf{r}_n^T\mathbf{A}\mathbf{r}_n}\qquad (\mathbf{r}_n=\mathbf{A}\mathbf{x}_n-\mathbf{b})$$

----

### 例
実験してみましょう。
$$ \mathbf{A} = \begin{pmatrix}2 & 1 \\ 1 & 4 \end{pmatrix}, \mathbf{b}=\begin{pmatrix} 1 \\ 2\end{pmatrix}, c = 0$$
の場合を考えます。つまり

$$f(x,y) = \frac{1}{2}\begin{pmatrix} x & y \end{pmatrix}\begin{pmatrix}2 & 1 \\ 1 & 4 \end{pmatrix}\begin{pmatrix}x\\y\end{pmatrix}-\begin{pmatrix} 1 & 2 \end{pmatrix}\begin{pmatrix} x \\ y \end{pmatrix} = \frac{1}{2}(2x^2+2xy+4y^2)-(x+2y)=x^2+xy+2y^2-x-2y$$

という関数を最小化します。2変数の場合は簡単に厳密解を求める事が可能です。この場合

$$f(x,y) = x^2 + xy + 2y^2 -x-2y = \left(x + \frac{y-1}{2}\right)^2 + \frac{7}{4}\left(y - \frac{3}{7}\right)^2 - \frac{4}{7} $$

なので、これが最小となるのは $(x, y) = (2/7, 3/7) = (0.285714\cdots, 0.428571\cdots)$ となります。

多変数の場合にはこの方法では計算量が増加してしまう為、勾配法などで近似解を求める事が必要になります。
初期値を $(0,0)$ として実験してみましょう。

```python
# ステップサイズ固定の場合
ALPHA = 0.1
EPS   = 1.0e-5
A = np.array([[2, 1], [1, 4]], dtype=float)
b = np.array([1, 2], dtype=float)
x = np.zeros(2)
for k in range(1000):
    x_new = x - ALPHA*(A.dot(x)-b)
    if np.linalg.norm(x_new - x) < EPS: break
    x = x_new
print ("count=",k+1)
print (x)
```

ステップサイズ固定 ($\alpha=0.1$)の場合は約40回で求める解に収束しました。次は理想的なステップサイズを計算する場合です。

```python
# 最適なステップサイズを使う場合
EPS   = 1.0e-5
A = np.array([[2, 1], [1, 4]], dtype=float)
b = np.array([1, 2], dtype=float)
x = np.array([0, 0], dtype=float)
for k in range(1000):
    r = A.dot(x)-b
    ALPHA = r.dot(r)/r.dot(A.dot(r))
    x_new = x - ALPHA*r
    if np.linalg.norm(x_new - x) < EPS: break
    x = x_new
print ("count=",k+1)
print (x)
```

今度は6回で収束しました。上と比べて極めて効率的である事が判ると思います。等高線を描いてみると、下図のようになり確かに一歩毎にその方向で最小の地点まで進んでいる事が分かると思います。またこの時、$\mathbf{x}_n$は直角に交わるギザギザな経路を進むことに注目しましょう。

```python
X, Y = meshgrid(linspace(0, 0.5, 20), linspace(0, 0.5, 20))
xlim(0, 0.5)
ylim(0, 0.5)
Z = X**2 + X*Y + 2*Y**2 - X - 2*Y
contour(X, Y, Z,50)
EPS   = 1.0e-5
A = np.array([[2, 1], [1, 4]], dtype=float)
b = np.array([1, 2], dtype=float)
x = np.array([0, 0], dtype=float)
traj = []
for k in range(1000):
    traj.append(x)
    r = A.dot(x)-b
    ALPHA = r.dot(r)/r.dot(A.dot(r))
    x_new = x - ALPHA*r
    if np.linalg.norm(x_new - x) < EPS: break
    x = x_new
traj = np.array(traj)
scatter(traj[:,0], traj[:,1], s=30, c="red")
for k in range(1,len(traj)):
    p0 = traj[k-1, :]
    p1 = traj[k, :]
    d  = p1 - p0
    arrow(p0[0], p0[1], d[0], d[1], head_width=0.01, length_includes_head=True)
axes().set_aspect('equal')
```

## 共役勾配法

先ほどの二次計画問題においてステップサイズを厳密に計算する方法では，経路が直角にギザギザしたものになるので収束が遅くなってしまいます。このように再急降下法は必ずしも良い方法論ではありません。二次計画問題に対しては**共役勾配法(conjugate gradient method)**という方法が有用です。これは変数が $N$ 個の場合、理論上は確実に $N$ ステップ以下で真の解に収束するという著しい特徴を持っています。(実際には演算誤差が生じる為$N$ステップより若干増える場合もあります。)

共役勾配法は図形的に理解するよりも、線型代数を使って理解した方が分かりやすいと思います。
先ほどと同じく以下の形の関数の最小化を考えます。ベクトルの次元は $N$ とします。$\mathbf{A}$ は正定値対称行列です。
$$f(\mathbf{x}) = \frac{1}{2}\mathbf{x}^T\mathbf{A}\mathbf{x} - \mathbf{b}^T\mathbf{x} + c $$
$f(\mathbf{x})$ が最小になる $\mathbf{x}$ を $\mathbf{x}^*$ とすると，この点で勾配が $0$ になるので
$$ \nabla f(\mathbf{x}^*) = \mathbf{A}\mathbf{x}^*-\mathbf{b} = 0\qquad\cdots (1)$$
が成り立ちます。

２つのベクトル $\mathbf{p}_i,\mathbf{p}_j$ が
$$ \mathbf{p}_i^T\mathbf{A}\mathbf{p}_j = 0$$
を満たすとき、２つのベクトルは$\mathbf{A}$に関して**共役(conjugate)**であると言います。共役勾配法では進行方向 $\mathbf{p}_n$ が全て互いに共役になるように選択します。また、ステップ幅パラメータ $\alpha_n$ は命題2.1を使って最適な値を選びます。つまり、以下のようなアルゴリズムです。

---- 

1. $\mathbf{x}_1 = 0$ とする。
2. $n=1,\ldots,N$について以下を繰り返す。
    1. $\mathbf{p}_1,\ldots,\mathbf{p}_{n-1}$の全てと共役な$\mathbf{p}_n$を選ぶ。
    2. その方向の最適なステップ幅パラメータ $\alpha_n$ を命題2.1によって求める。
    3. $\mathbf{x}_{n+1} = \mathbf{x}_n + \alpha_n\mathbf{p}_n$ によって更新する。

----

$\alpha_n$ の具体的な立式を求める前に、幾つか確認しましょう。まず、
$$ \mathbf{x}_{n} = \alpha_1\mathbf{p}_1 + \alpha_2\mathbf{p}_2 + \cdots + \alpha_n\mathbf{p}_{n-1} $$
が成り立ちます。従って
$$ \mathbf{p}_n^T\mathbf{A}\mathbf{x}_n = \alpha_1\mathbf{p}_n^T\mathbf{A}\mathbf{p}_1+\alpha_2\mathbf{p}_n^T\mathbf{A}\mathbf{p}_2+\cdots+\alpha_n\mathbf{p}_n^T\mathbf{A}\mathbf{p}_{n-1}$$
となりますが、$\{\mathbf{p}_1,\mathbf{p}_2,\ldots,\mathbf{p}_n\}$は互いに共役なのでこれは $0$ になります。

----
【命題2.2】

共役勾配法のアルゴリズムにおいて $\mathbf{p}_n^T\mathbf{A}\mathbf{x}_n = 0$ が成り立つ。

----

従って命題2.1より
$$\alpha_n = -\frac{\mathbf{p}_n^T(\mathbf{A}\mathbf{x}_n-\mathbf{b})}{\mathbf{p}_n^T\mathbf{A}\mathbf{p}_n} = \color{red}{ \frac{\mathbf{p}_n^T\mathbf{b}}{\mathbf{p}_n^T\mathbf{A}\mathbf{p}_n} }$$
である事が分かります。よって

$$\mathbf{x}_{N+1} = \frac{\mathbf{p}_1^T\mathbf{b}}{\mathbf{p}_1^T\mathbf{A}\mathbf{p}_1}\mathbf{p}_1+\cdots+\frac{\mathbf{p}_N^T\mathbf{b}}{\mathbf{p}_N^T\mathbf{A}\mathbf{p}_N}\mathbf{p}_N$$

となりますが、実は $\mathbf{x}^*=\mathbf{x}_{N+1}$になっています。

----
【命題.23】

共役勾配法のアルゴリズムにおいて $\mathbf{x}^*=\mathbf{x}_{N+1}$ である。

【証明】

$\{\mathbf{p}_1,\mathbf{p}_2,\ldots,\mathbf{p}_N\}$ は一次独立であるので、$\mathbb{R}^N$の基底をなす。従って
$$ \mathbf{x}^* = \beta_1\mathbf{p}_1 + \cdots \beta_N\mathbf{x}_N$$
と一意的に表す事が出来る。この両辺に $\mathbf{p}_n^T\mathbf{A}$ を掛けると、$\{\mathbf{p}_1,\ldots,\mathbf{p}_N\}$が互いに共役であることより
$$ \mathbf{p}_n^T\mathbf{A}\mathbf{x}^* = \beta_n\mathbf{p}_n^T\mathbf{A}\mathbf{p}_n$$
となる。これと(1)より
$$\beta_n = \frac{\mathbf{p}_n^T\mathbf{b}}{\mathbf{p}_n^T\mathbf{A}\mathbf{p}_n}$$
となり，$\alpha_n = \beta_n$ となる。従って $\mathbf{x}^*=\mathbf{x}_{N+1}$ である。

----

このことから、共役勾配法のアルゴリズムはちょうど $N$ ステップで終了する事が分かります。

残る問題は $\{\mathbf{p}_1,\cdots,\mathbf{p}_n\}$ を求める事のみですが、これには[シュミットの直交化法](http://ja.wikipedia.org/wiki/%E3%82%B0%E3%83%A9%E3%83%A0%E3%83%BB%E3%82%B7%E3%83%A5%E3%83%9F%E3%83%83%E3%83%88%E3%81%AE%E6%AD%A3%E8%A6%8F%E7%9B%B4%E4%BA%A4%E5%8C%96%E6%B3%95)を使う事が出来ます。シュミットの直交化法は適当なベクトルのセット $\{\mathbf{r}_1,\ldots,\mathbf{r}_n\}$ から直交する(今回の場合共役な)ベクトルのセットを導出する方法です。$\mathbf{r}_n$は何でも良いのですが、共役勾配法では普通

$$\mathbf{r}_n = -\nabla f(\mathbf{x}_n) = \mathbf{b}-A\mathbf{x}_n$$

とします。こうすると次の非常に良い性質が成り立ちます。

----
【命題2.4】
$$\mathbf{r}_n^T\mathbf{p}_k=0\quad (k=1,2,\ldots,n-1)$$


【証明】
$$\begin{align*}
\mathbf{r}_n^T\mathbf{p}_k &= \mathbf{b}^T\mathbf{p}_k-\mathbf{x}_n^T\mathbf{A}\mathbf{p}_k\\
                           &= \mathbf{b}^T\mathbf{p}_k-\alpha_k\mathbf{p}_k^T\mathbf{A}\mathbf{p}_k\\
                           &= \mathbf{b}^T\mathbf{p}_k-\mathbf{b}^T\mathbf{p}_k\\
                           &= 0
\end{align*}$$

----
この命題2.4に気をつけてシュミットの直交化法を使ってみましょう。

まず$\mathbf{x}_1=0$ なので $\mathbf{r}_1 = \mathbf{b}$ です。最初の基底ベクトルは単に $\color{red}{\mathbf{p}_1 = \mathbf{r}_1}$ とします。
次は$\mathbf{r}_2=A\mathbf{x}_2-\mathbf{b}$とし，

$$\color{red}{\mathbf{p}_2 = \mathbf{r}_2-\frac{\mathbf{p}_1^T\mathbf{A}\mathbf{r}_2}{\mathbf{p}_1^T\mathbf{A}\mathbf{p}_1}\mathbf{p}_1}$$

と置きます。これに左から$\mathbf{p}_1^T\mathbf{A}$を掛けると

$$\mathbf{p}_1^T\mathbf{A}\mathbf{p}_2 = \mathbf{p}_1^T\mathbf{A}\mathbf{r}_2 - \frac{\mathbf{p}_1^T\mathbf{A}\mathbf{r}_2}{\mathbf{p}_1^T\mathbf{A}\mathbf{p}_1}\mathbf{p}_1^T\mathbf{A}\mathbf{p}_1 = 0$$

となるので確かに$\mathbf{p}_1,\mathbf{p}_2$は互いに共役になっています。

次は $\mathbf{r}_3=A\mathbf{x}_3-\mathbf{b}$ とし

$$\mathbf{p}_3= \mathbf{r}_3-\frac{\mathbf{p}_2^T\mathbf{A}\mathbf{r}_3}{\mathbf{p}_2^T\mathbf{A}\mathbf{p}_2}\mathbf{p}_2-\frac{\mathbf{p}_1^T\mathbf{A}\mathbf{r}_3}{\mathbf{p}_1^T\mathbf{A}\mathbf{p}_1}\mathbf{p}_1\quad\cdots(4)$$

と置きます。この両辺に$\mathbf{p}_1^T\mathbf{A}$を掛けると

$$\mathbf{p}_1^T\mathbf{A}\mathbf{p}_3= \mathbf{p}_1^T\mathbf{A}\mathbf{r}_3-\frac{\mathbf{p}_2^T\mathbf{A}\mathbf{r}_3}{\mathbf{p}_2^T\mathbf{A}\mathbf{p}_2}\mathbf{p}_1^T\mathbf{A}\mathbf{p}_2-\frac{\mathbf{p}_1^T\mathbf{A}\mathbf{r}_3}{\mathbf{p}_1^T\mathbf{A}\mathbf{p}_1}\mathbf{p}_1^T\mathbf{A}\mathbf{p}_1$$

となりますが，$\mathbf{p}_1^T\mathbf{A}\mathbf{p}_2=0$なので$\mathbf{p}_1^T\mathbf{A}\mathbf{p}_3= 0$となります。同様に$\mathbf{p}_2^T\mathbf{A}\mathbf{p}_3=0$も成り立つので，$\{\mathbf{p}_1,\mathbf{p}_2,\mathbf{p}_3\}$は互いに共役となります。
この時$(4)$式において命題2.3を使うと

$$\mathbf{p}_1^T\mathbf{A}\mathbf{r}_3=\mathbf{r}_3^T\mathbf{A}\mathbf{p}_1=\mathbf{r}_3^T\mathbf{A}\left(\frac{\mathbf{x}_2-\mathbf{x}_1}{\alpha_1}\right)=\frac{1}{\alpha_1}\mathbf{r}_3^T(\mathbf{r}_1-\mathbf{r}_2) = 0 \qquad (\because \mathbf{r}_1-\mathbf{r}_2\in\mathrm{span}\{\mathbf{p}_1,\mathbf{p}_2\})$$

が成り立つので
$$\color{red}{ \mathbf{p}_3= \mathbf{r}_3-\frac{\mathbf{p}_2^T\mathbf{A}\mathbf{r}_3}{\mathbf{p}_2^T\mathbf{A}\mathbf{p}_2}\mathbf{p}_2 }$$

と簡単に書くことが出来ます。一般に
$$\color{red}{ \mathbf{p}_{n+1}= \mathbf{r}_{n+1}-\frac{\mathbf{p}_n^T\mathbf{A}\mathbf{r}_{n+1}}{\mathbf{p}_n^T\mathbf{A}\mathbf{p}_n}\mathbf{p}_n }$$
とすれば良いです。

以上で共役勾配法のアルゴリズムが完成しました。

----

1. $\mathbf{x}_1 = 0$
2. $\mathbf{r}_1 = \mathbf{b}$
3. $\mathbf{p}_1 = \mathbf{r}_1$
4. $n=1,\ldots$について以下を繰り返す。
    1. $\alpha_n = \frac{\mathbf{p}_n^T\mathbf{b}}{\mathbf{p}_n^T\mathbf{A}\mathbf{p}_n}$
    2. $\mathbf{x}_{n+1} = \mathbf{x}_n + \alpha_n\mathbf{p}_n$
    3. $|\mathbf{x}_{n+1}-\mathbf{x}_n|<\varepsilon$ならば終了。
    4. $\mathbf{r}_{n+1} = \mathbf{r}_r - \alpha_n\mathbf{A}\mathbf{p}_n$
    5. $\mathbf{p}_{n+1} = \mathbf{r}_{n+1} - \frac{\mathbf{p}_n^T\mathbf{A}\mathbf{r}_{n+1}}{\mathbf{p}_n^T\mathbf{A}\mathbf{p}_n}\mathbf{p}_n$


先ほどの以下の問題に適用してみましょう。最急降下法＋最適なステップ幅選択で初期値を$(1,1)$にした場合、9ステップ必要でした。
$$ \mathbf{A} = \begin{pmatrix}2 & 1 \\ 1 & 4 \end{pmatrix}, \mathbf{b}=\begin{pmatrix} 1 \\ 2\end{pmatrix}, c = 0$$

```python
EPS   = 1.0e-5
A = np.array([[2, 1], [1, 4]], dtype=float)
b = np.array([1, 2], dtype=float)
x = np.zeros(2)
r = b
p = r
for k in range(1000):
    pAp = p.dot(A.dot(p))
    ALPHA = p.dot(b)/pAp
    x_new = x + ALPHA*p
    if np.linalg.norm(x_new - x) < EPS: break
    x  = x_new
    r -= ALPHA*A.dot(p)
    p  = r - p.dot(A.dot(r))/pAp * p
print ("count=",k+1)
print (x)
```

```python
X, Y = meshgrid(linspace(-0.5, 0.5, 20), linspace(-0.5, 0.5, 20))
xlim(-0.5, 0.5)
ylim(-0.5, 0.5)
Z = X**2 + X*Y + 2*Y**2 - X - 2*Y
contour(X, Y, Z,50)
EPS = 1.0e-5
A = np.array([[2, 1], [1, 4]], dtype=float)
b = np.array([1, 2], dtype=float)
x = np.zeros(2)
r = b
p = r
traj = []
for k in range(2):
    traj.append(x)
    pAp = p.dot(A.dot(p))
    ALPHA = p.dot(b)/pAp
    x_new = x + ALPHA*p
    if np.linalg.norm(x_new - x) < EPS: break
    x  = x_new
    r -= ALPHA*A.dot(p)
    p  = r - p.dot(A.dot(r))/pAp * p
traj.append(x)
traj = np.array(traj)
scatter(traj[:,0], traj[:,1], s=30, c="red")
for k in range(1,len(traj)):
    p0 = traj[k-1, :]
    p1 = traj[k, :]
    d  = p1 - p0
    arrow(p0[0], p0[1], d[0], d[1], head_width=0.03, length_includes_head=True)
axes().set_aspect('equal')
```

## 二次計画問題に対する勾配法の計算量について
何故、二次計画問題に勾配法を用いる事を考えるのか疑問を感じたかもしれません。何故ならば

$$f(\mathbf{x}) = \frac{1}{2}\mathbf{x}^T\mathbf{A}\mathbf{x} - \mathbf{b}^T\mathbf{x} + c $$

が最小となるのは

$$ \nabla f(\mathbf{x}) = \mathbf{A}\mathbf{x} - \mathbf{b} $$

が $\mathbf{0}$ になる時なので、

$$ \mathbf{A}\mathbf{x} = \mathbf{b}\Leftrightarrow \mathbf{x} = \mathbf{A}^{-1}\mathbf{b}$$

と直接解を求めてしまう事が可能であるからです。このような解法を**直接法(direct method)**と呼びます。

この理由は計算量にあります。変数の数を $n$ とすると方程式 $\mathbf{A}\mathbf{x}=\mathbf{b}$ を解くためには一般に $\mathcal{O}(n^3)$の計算時間が必要になります。例えば、変数の数が $10$ 倍になると大体 $1000$ 倍の計算時間が必要になります。

勾配法のアルゴリズムでは一ステップの更新に $\mathcal{O}(n^2)$ の計算時間が必要です。これと反復回数の積が実際の計算時間になるわけですが、収束するまでの反復回数が $n$ よりも十分に少なければ総計算量は $\mathcal{O}(n^3)$ よりも少なくなります。実際に実験をしてみましょう。

まず、ランダムに正定値対称行列 $\mathbf{A}$ とベクトル $\mathbf{b}$ を容易します。変数は3000個としました。

```python
from time import time
N = 3000 # 変数の数
A = np.random.uniform(0, 1, (N, N)) # ランダムなNxN行列
A += A.T         # 対称行列に加工
A += N*np.eye(N) # 正定値行列に加工
b = np.random.uniform(0, 1, N) # ランダムなN次元ベクトル
```


以下が直接法で解いた場合です。約2秒掛かっています。

```python
start = time()
x = np.linalg.solve(A, b)  # Ax = b を解く
end = time()
print (end-start)
print (x)
```

以下が共役勾配法で解いた場合です。0.1秒未満で解を求める事ができています。

```python
start = time()
EPS   = 1.0e-5
x = np.zeros(N)
r = b
p = r
for k in range(1000):
    pAp = p.dot(A.dot(p))
    ALPHA = p.dot(b)/pAp
    x_new = x + ALPHA*p
    if np.linalg.norm(x_new - x) < EPS: break
    x  = x_new
    r -= ALPHA*A.dot(p)
    p  = r - p.dot(A.dot(r))/pAp * p
end = time()
print ("count=",k+1)
print (end-start)
print (x)
```
