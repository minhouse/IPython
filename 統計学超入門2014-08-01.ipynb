平均
$$ \bar{x} = \frac{1}{2}\sum_{k=1}^n x_k$$

```python
average(x)
```

二乗和
$$ \sum_{k=1}^nx_k^2$$

```python
sum(x**2)
```

線形回帰式 $y=ax+b$ の係数を最小二乗法で決定する場合

$$ a = \frac{\sum x_ky_k - \bar{y}\sum x_k}{\sum x_k^2 - \bar{x}\sum x_k},\quad b = \bar{y} - \bar{x}a $$

```python
a = (sum(x*y) - average(y) * sum(x))/(sum(x**2) - average(x) * sum(x))
b = average(y) - average(x) * a
a, b
```

## モデルの複雑さと汎化性能について
モデルの複雑さと汎化性能にはトレードオフがある。
「モデル」というのは「実際に観測された値」から学習される。
学習に用いたデータに対する当てはまりの良さだけを考えては行けない。
大事なのは学習に使われていないデータに対する当てはまりの良さ(汎化性能)。

```python
N = 10
x = random.uniform(0, 100, N)
y = random.normal(2*x + 3, 30)
scatter(x, y, s=10)
```

これに対して例えば
$$ y = a_dx^d + a_{d-1}x^{d-1}+\cdots + a_1x + a_0$$
という多項式モデルを当てはめてみたいとします。$d$ が大きくなるほど、モデルは複雑になる。

```python
d = 1
X = array([x**k for k in range(d+1)]).T
a = linalg.solve(X.T.dot(X), X.T.dot(y))
test_x = linspace(0, 100)
test_y = array([test_x**k for k in range(d+1)]).T.dot(a)
scatter(x, y, s=10)
plot(test_x, test_y)
```

```python
d = 2
X = array([x**k for k in range(d+1)]).T
a = linalg.solve(X.T.dot(X), X.T.dot(y))
test_x = linspace(0, 100)
test_y = array([test_x**k for k in range(d+1)]).T.dot(a)
scatter(x, y, s=10)
plot(test_x, test_y)
```

```python
d = 3
X = array([x**k for k in range(d+1)]).T
a = linalg.solve(X.T.dot(X), X.T.dot(y))
test_x = linspace(0, 100)
test_y = array([test_x**k for k in range(d+1)]).T.dot(a)
scatter(x, y, s=10)
plot(test_x, test_y)
```

上の1次～３次のモデルでは、**学習データに対する誤差**が存在するが、未知のデータに対してもある程度当てはまる事が期待できる。汎化性能は悪くない。

```python
d = 10
X = array([x**k for k in range(d+1)]).T
a = linalg.solve(X.T.dot(X), X.T.dot(y))
test_x = linspace(0, 100)
test_y = array([test_x**k for k in range(d+1)]).T.dot(a)
scatter(x, y, s=10)
plot(test_x, test_y)
```

上の 10次モデルは学習データに対する誤差は ** 0 **だが、未知のデータに対する当てはまりは非常に悪い。汎化性能が低い。過学習とかオーバーフィッティングと呼ばれる現象。

```python
N = 1000
x = random.uniform(0, 100, N)
y = random.normal(2*x + 3, 30)
scatter(x, y, s=10)
```

```python
d = 10
X = array([x**k for k in range(d+1)]).T
a = linalg.solve(X.T.dot(X), X.T.dot(y))
test_x = linspace(0, 100)
test_y = array([test_x**k for k in range(d+1)]).T.dot(a)
scatter(x, y, s=10)
plot(test_x, test_y, c="red")
```
