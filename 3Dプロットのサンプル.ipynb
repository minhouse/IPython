3Dグラフ表示のテストです。matplotlibのAxes3Dというモジュールをインポートします。3Dの軸を管理するモジュールです。

```python
from mpl_toolkits.mplot3d import Axes3D
```

新しいキャンバスを関数figure()で作り、その上に3次元軸を構築します。

```python
fig = figure()
ax = Axes3D(fig)
```

グリッドを作ります。$x,y$ 軸共に $[-1, 1]$ の範囲を $20$ 分割しています。

```python
X, Y = meshgrid(linspace(-1, 1, 20), linspace(-1, 1, 20))
```

$z = x^2 + y^2$ という放物曲面を描いてみます。上で求めた各 $(x,y)$ に対して $z$ の値を計算します。

```python
Z = X**2 + Y**2
```

ワイヤーフレームで描画してみます。

```python
ax.plot_wireframe(X, Y, Z)
fig
```

見る方向を変えるにはAxes3Dオブジェクトのview_initメソッドを使います。elevで上下方向の角度、azimで横方向の角度を与えます。
例えば真横から見るならば elev = 0にすれば良いです。

```python
ax.view_init(elev=0)
fig
```

以下は等高線を描く例です。他の様々な例はhttp://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html などを参照。

```python
ax.contourf(X, Y, Z, offset=0)
ax.view_init(elev=60, azim=30)
fig
```
